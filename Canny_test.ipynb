```python
import cv2
import matplotlib.pyplot as plt

# Read the image
img = cv2.imread('rs00003.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply Gaussian blur to reduce noise
blurred_img = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
# Apply Canny edge detection
edges = cv2.Canny(blurred_img, 50, 150)
```

```python
# Show the edge-detected image
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.show()
```

```python
import numpy as np
lines = cv2.HoughLinesP(edges, rho=1, theta=np.pi / 180, threshold=100, minLineLength=100, maxLineGap=10)

# Draw the lines on the original image
line_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(line_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Display the result with lines
plt.imshow(line_img)
plt.title('Railroad Tracks Detection with Hough Lines')
plt.show()
```
